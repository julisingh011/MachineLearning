```python
import pandas as pd 
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=1000,n_classes=2,random_state=1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
dummy_mod=[0 for i in range(len(x_test))]
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model_prob=model.predict_proba(x_test)
model_prob=model_prob[:,1]
model_prob
```

```python
from sklearn.metrics import roc_curve,roc_auc_score

dummy_scr=roc_auc_score(y_test,dummy_mod)
model_scr=roc_auc_score(y_test,model_prob)
print(dummy_scr)
print(model_scr)
```

```python
dummy_fpr,dummy_tpr,threshold=roc_curve(y_test,dummy_mod)
model_fpr,model_tpr,threshold=roc_curve(y_test,model_prob)
```

```python
import matplotlib.pyplot as plt

plt.plot(dummy_fpr,dummy_tpr,linestyle="--",label='Dummy Model')
plt.plot(model_fpr,model_tpr,marker=',',label='Logistic Regression')
         
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
model_prob=model.predict_proba(x_test)
model_prob=model_prob[:,1]
model_prob
```

```python
import matplotlib.pyplot as plt

plt.plot(dummy_fpr,dummy_tpr,linestyle="--",label='Dummy Model')
plt.plot(model_fpr,model_tpr,marker=',' ,label='Logistic Regression')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
for i ,thr in enumerate(threshold):
   if i%5==0:
       plt.annotate(f"(thr:.1f)",(model_fpr[i],model_tpr[i]))
       
plt.legend()
plt.show()
```
