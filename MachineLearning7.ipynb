```python
import numpy as np
x=np.linspace(-5,5,100)
y=2*x**2+3*x+5+np.random.normal(0,5,100)
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt 

plt.scatter(x,y)
plt.xlabel("x label")
plt.ylabel("Target")
plt.legend()
plt.title("synthetic data")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error 
model=LinearRegression()
model.fit(x.reshape(-1,1),y)
y_pred_linear=model.predict(x.reshape(-1,1))
poly =PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x.reshape(-1,1))
x_poly
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error 
model=LinearRegression()
model.fit(x.reshape(-1,1),y)
y_pred_linear=model.predict(x.reshape(-1,1))
poly =PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x.reshape(-1,1))
# x_poly
polymod=LinearRegression()
polymod.fit(x_poly,y)
y_pred_poly=polymod.predict(x_poly)
print(mean_squared_error(y,y_pred_linear))
print(mean_squared_error(y,y_pred_poly))
```

```python
plt.scatter(x,y,color="red",label="ActualDataPoints")
plt.plot(x,y_pred_linear,color="blue",label="LinearReg")
plt.plot(x,y_pred_poly,color="green",label="PolyReg")
plt.legend()
plt.show()
```
