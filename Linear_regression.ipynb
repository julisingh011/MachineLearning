# ***LINEAR REGRESSION***

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# data={'area':[1000,1200,1500,2000,2300,2500,3000,3500],
#       'price':[4500000,5000000,5500000,6500000,7000000,7550000,8500000,9500000]}
```

```python
df=pd.read_csv('/home/praveen/Desktop/DATA ANALYSIS CS/Machine Learning/areaprice.csv')
df
```

```python
plt.plot(df.area,df.price,marker='+')
plt.show()
```

```python
dfa=df.drop('price',axis=1)
dfa
```

```python
lreg=LinearRegression()
lreg.fit(dfa.values,df['price'].values)
```

```python
lreg.predict([[2200]])
```

```python
m=lreg.coef_
c=lreg.intercept_
print(m)
print(c)
```

```python
yhat=m*2200+c
print(yhat)
```

```python
df_areas={'areas':[700,1150,2200,2650,2500,2800,3200]}
dfar=pd.DataFrame(df_areas)
dfar
```

```python
pred=lreg.predict(dfar.values)
dfar['Predictions']=pred
dfar
```
